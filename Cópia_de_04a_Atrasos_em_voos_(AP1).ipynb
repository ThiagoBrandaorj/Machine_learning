## Informações incluídas

- **Detalhes do Voo**: informações sobre data e hora do voo, companhia aérea, número do voo, aeroportos de origem e destino.

- **Status Operacional**: indicadores que indicam se um voo foi concluído, atrasado, cancelado ou desviado.

- **Métricas de Atraso**: atraso total em minutos, bem como detalhamentos por causas específicas, como problemas com a companhia aérea, clima, atrasos no Sistema Nacional de Espaço Aéreo (NAS), segurança e aeronaves com atraso.

- **Motivos de Cancelamento**: motivos categóricos para cancelamentos de voos.

- **Intervalos de Tempo**: valores de tempo local para eventos programados e reais (partidas e chegadas).

O conjunto de dados foi compilado agregando relatórios mensais do Departamento de Transportes (DOT) e passou por limpeza e transformação para facilitar a análise estruturada.

## Variáveis

- **FL_DATE**: Data do voo.
- **AIRLINE**: Nome da companhia aérea.

- **FL_NUMBER**: Número do voo.

- **TAXI_OUT**: Tempo gasto no taxiamento.

- **TAXI_IN**: Tempo gasto taxiando para chegar.
- **CRS_ARR_TIME**: Horário de chegada programado.
- **ARR_TIME**: Horário de chegada real.

- **CANCELLED**: Indicador se o voo foi cancelado (1 para cancelado, 0 para não cancelado).

- **CRS_ELAPSED_TIME**: Tempo decorrido programado.
- **ELAPSED_TIME**: Tempo decorrido real.
- **AIR_TIME**: Tempo gasto no ar.
- **DISTANCE**: Distância percorrida.
- **DELAY_DUE_CARRIER**: Atraso devido à companhia aérea.
- **DELAY_DUE_WEATHER**: Atraso devido ao clima.
- **DELAY_DUE_NAS**: Atraso devido ao Sistema Nacional de Espaço Aéreo (NAS).
- **DELAY_DUE_SECURITY**: Atraso devido à segurança.

````python
# lista com as colunas que representam diferentes causas de atraso
delay_cols = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

mask_arr_neg = df['ARR_DELAY'] <= 0
mask_arr_pos = df['ARR_DELAY'] > 0

# percorre cada coluna de causa de atraso da lista
for col in delay_cols:
    print(f"\n── {col} ──")
    for label, mask in [('ARR_DELAY <= 0', mask_arr_neg), ('ARR_DELAY > 0', mask_arr_pos)]:
        total = mask.sum()
        na_count = df.loc[mask, col].isna().sum()
        print(f"  {label}: {na_count}/{total} NAs ({na_count/total:.1%})")
````

**Analisando melhor as variáveis mencionadas, além de conterem muitos NAs, as variáveis possuem dados que não fazem sentido, considerando que deveriam ser variáveis de flag. Logo, iremos excluí-las.**

````python
# converte o horário real de partida em períodos do dia (manhã, tarde ou noite), cria a variável 'DEP_PERIOD' e remove a coluna original 'DEP_TIME'
def hora_para_periodo(x):
    if pd.isna(x):
        return None
    hora = int(str(int(x)).zfill(4)[:2])
    if 6 <= hora < 12:
        return "manha"
    elif 12 <= hora < 18:
        return "tarde"
    else:
        return "noite"

df["DEP_PERIOD"] = df["DEP_TIME"].apply(hora_para_periodo)
df = df.drop(columns=['DEP_TIME'])
````

````python
# Convertendo 'ano' e 'trimestre' para categóricas para serem tratadas pelo OneHotEncoder no pré-processamento.

df['ano'] = df['ano'].astype(object)
df['trimestre'] = df['trimestre'].astype(object)

# Garantindo que 'CRS_ARR_TIME' esteja em formato numérico contínuo para que possa ser escalada corretamente pelo RobustScaler.
df['CRS_ARR_TIME'] = df['CRS_ARR_TIME'].astype(float)
````

````python
# Separando as variáveis explicativas (X) da variável alvo (y), que indica se o voo teve atraso de chegada de pelo menos 15 minutos.

X = df.drop('ARR_DELAY_FLAG',axis=1)
y = df['ARR_DELAY_FLAG']
````

````python
from sklearn.model_selection import train_test_split

# Split inicial: treino+validação vs teste (20% teste)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=df['ARR_DELAY_FLAG'],
    random_state=42
)

#  Split: treino vs validação (dos 80% restantes)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=0.25,  # 0.25 * 0.8 = 0.20 → 20% do total → 15% do total final
    stratify=y_temp,
    random_state=42
)

# Tamanhos
print(f"Treino: {X_train.shape[0]}")
print(f"Validação: {X_val.shape[0]}")
print(f"Teste: {X_test.shape[0]}")

# Proporção da classe positiva
y_train.mean(), y_val.mean(), y_test.mean()
````

````python
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
import pandas as pd

# Variáveis categóricas que serão transformadas em representação numérica por meio de One-Hot Encoding.
cat_features = ["AIRLINE", "ORIGIN", "DEST","ano","trimestre","DEP_PERIOD"]

# Variáveis numéricas contínuas que serão escaladas com RobustScaler para reduzir o impacto de outliers.
numeric_features = ["DEP_DELAY","TAXI_OUT","WHEELS_OFF","CRS_ARR_TIME","CRS_ELAPSED_TIME","sin_mes","sin_dia","cos_mes","cos_dia"]

# Variáveis binárias que já estão em formato adequado e, por isso, serão mantidas sem transformação.
no_modified_features = ["DIVERTED","is_fim_semana"]

# Pré-processamento
preprocess = ColumnTransformer(
    transformers=[
        (
            "cat",
            OneHotEncoder(
                drop="first", # Remove a primeira categoria de cada variável categórica, reduzindo a redundância
                handle_unknown="ignore", # Ignora categorias não vistas no treino, evitando erro ao transformar os conjuntos de validação e teste
                sparse_output=False
            ),
            cat_features,
        ),
        (
            "num",
            RobustScaler(),
            numeric_features,
        ),
    ],
    remainder="passthrough"
)

# Aplicar transformação
X_train_array = preprocess.fit_transform(X_train)
X_test_array  = preprocess.transform(X_test)

# Recuperar nomes das features
cat_feature_names = preprocess.named_transformers_["cat"].get_feature_names_out(cat_features)

feature_names = (
    list(cat_feature_names)
    + numeric_features
    + no_modified_features
)

# Criar DataFrame final
X_train_processed = pd.DataFrame(X_train_array, columns=feature_names, index=X_train.index)
````

````python
# Transformando o conjunto de validação com o mesmo pré-processamento aprendido a partir do conjunto de treino.
X_val_array = preprocess.transform(X_val)
````

# 3 - MLP

````python
# TensorFlow será utilizado para construir e treinar a rede neural MLP.
import tensorflow as tf

# Definindo uma rede neural para classificação binária de voos atrasados e não atrasados.
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(795,)), # Camada de entrada com 795 atributos, correspondente ao número de features geradas após o pré-processamento.
    tf.keras.layers.Dense(64, activation="relu"), # Camadas densas com função de ativação ReLU, responsáveis por aprender relações não lineares entre as variáveis de entrada.
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(1, activation='sigmoid') # Camada de saída com ativação sigmoide, retornando a probabilidade de o voo pertencer à classe positiva (atraso >= 15 minutos).
])

model.compile(
    optimizer='adam', # Adam como otimizador
    loss='binary_crossentropy', # binary_crossentropy como função de perda
    metrics=[tf.keras.metrics.Recall()] # recall como métrica principal, devido ao desbalanceamento das classes
)
model.summary()
````

**Foi utilizada a função de ativação ReLU nas camadas ocultas, por sua simplicidade computacional e bom desempenho em redes profundas.**

````python
#X_train = X_train_array.to_numpy(dtype="float32")
#X_val   = X_val_array.to_numpy(dtype="float32")
#X_test  = X_test_array.to_numpy(dtype="float32")
y_train = y_train.to_numpy()
y_val   = y_val.to_numpy()
y_test  = y_test.to_numpy()
````

````python
classes = np.unique(y_train)
from sklearn.utils.class_weight import compute_class_weight

# Calculando pesos para as classes de forma balanceada, dando mais importância à classe minoritária durante o treinamento.
weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=y_train
)
class_weight = dict(zip(classes, weights)) # Organizando os pesos em dicionário
class_weight
````

````python
history = model.fit(X_train_array,y_train,epochs=30,class_weight=class_weight,validation_data=(X_val_array,y_val))
````

````python
# Evolução das métricas e da função de perda ao longo das épocas
plt.figure(figsize=(8, 5))
for key, style in zip(history.history, ["r--", "r--.", "b-", "b-*"]):
    epochs = np.array(history.epoch) + (0 if key.startswith("val_") else -0.5)
    plt.plot(epochs, history.history[key], style, label=key)
plt.xlabel("Epoch")
plt.axis([-0.5, 29, 0., 1])
plt.legend(loc="lower left")
plt.grid()
plt.show()
````

````python
# Avaliando o modelo no conjunto de teste, obtendo a perda final e o recall fora da amostra.
test_loss,test_recall = model.evaluate(X_test_array,y_test)
print('test loss (binary_crossentropy):', test_loss, 'teste Recall:', test_recall)
````

````python
from sklearn.metrics import classification_report

print("\n" + "="*50)
print("CLASSIFICATION REPORT: (TREINO)")
print("="*50)
y_pred_prob = model.predict(X_train_array)
y_pred = (y_pred_prob > 0.4).astype(int)
print(classification_report(y_train, y_pred,
                           target_names=['Classe 0', 'Classe 1']))
print("\n" + "="*50)
print("CLASSIFICATION REPORT: (VALIDAÇÃO)")
print("="*50)
y_pred_prob = model.predict(X_val_array)
y_pred = (y_pred_prob > 0.4).astype(int) # Convertendo probabilidades em classes usando limiar 0.4,escolhendo uma regra mais sensível do que o padrão 0.5 para favorecer o recall.
print(classification_report(y_val, y_pred,
                           target_names=['Classe 0', 'Classe 1']))
print("\n" + "="*50)
print("CLASSIFICATION REPORT: (TESTE)")
print("="*50)
y_pred_prob = model.predict(X_test_array)
y_pred = (y_pred_prob > 0.4).astype(int)
print(classification_report(y_test, y_pred,
                           target_names=['Classe 0', 'Classe 1']))
````
